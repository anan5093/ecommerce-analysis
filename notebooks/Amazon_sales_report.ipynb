## 📊 Core Sales Performance KPIs (Post Data Cleaning)

After removing cancelled, returned, zero-revenue and incomplete transactions, the dataset reflects **net realized sales performance.**

### 💰 Total Revenue

- **₹69,663,293**

This indicates a strong overall transaction value generated during the observed period.  
The revenue scale suggests consistent marketplace demand and healthy order flow across product categories.

---

### 🧾 Total Orders

- **97,965 fulfilled orders**

This highlights significant platform activity and confirms that the dataset contains sufficient transaction volume for reliable trend, category, and behavioural analysis.

---

### 🛒 Average Order Value (AOV)

- **₹711.10**

The relatively moderate AOV suggests that:

- Customers are primarily purchasing mid-priced products  
- Demand may be driven by price-sensitive segments  
- Bulk premium purchases are not the dominant revenue driver  

This aligns with marketplace behaviour where high order frequency compensates for lower basket value.

---

### 📌 Key Analytical Implication

With revenue scale and AOV established, the next analytical step is to evaluate:

- **Revenue contribution by category**
- **Unit demand vs revenue gap**
- **Customer purchase behaviour patterns**

These insights will help identify pricing opportunities, inventory priorities, and growth drivers.

⭐ Revenue Share by Category

```python
category_rev = (
    df_net.groupby("Category")["Amount"].sum().sort_values(ascending=False)
)
category_rev
```

```python
plt.figure(figsize=(12,6))
category_rev.plot(kind="bar")

plt.title("Categories by Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```
