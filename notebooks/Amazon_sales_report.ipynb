KPI — Revenue Growth Rate
🎯 Business Meaning

Revenue Growth Rate measures how fast sales are increasing or decreasing over time.

It helps understand:

business momentum

seasonal spikes

demand expansion or contraction

This KPI is defined as:-

Revenue Growth Rate = (Current Period Revenue − Previous Period Revenue) / Previous Period Revenue.

````python
df_net["Date"] = pd.to_datetime(df_net["Date"])
````

````python
monthly_revenue = (
    df_net.groupby(df_net["Date"].dt.to_period("M"))["Amount"]
    .sum()
    .sort_index()
)

print("Monthly Revenue is:",monthly_revenue)
````

````python
revenue_growth = monthly_revenue.pct_change()
print("Revenue Growth is:", revenue_growth)
````

````python
plt.figure(figsize=(12,5))
revenue_growth.plot(marker="o")

plt.title("Monthly Revenue Growth Rate")
plt.ylabel("Growth Rate")
plt.xlabel("Month")
plt.grid()
plt.show()
````

## 🎯 Discount Effectiveness Analysis — Demand Impact

Such behaviour is common in marketplace-driven e-commerce environments where discounts are used to maintain competitive positioning and attract price-sensitive customers.

### 🛒 Impact of Promotions on Basket Size

Average quantity per order was analyzed for discounted vs non-discounted transactions:

| Promotion Status | Average Quantity per Order |

### 📌 Key Business Insight

````python
total_units = df_net["Qty"].sum()

units_per_transaction = (total_units / total_orders).round(3)

print("Unit Per Transaction:", units_per_transaction)
````

## 🛒 Units per Transaction Analysis

To understand customer basket behaviour, the average number of items purchased per transaction was calculated.

### 📊 Result

- **Units per Transaction = 1.077**

This indicates that customers typically purchase just over one item per order.

---

### 📌 Key Business Insight

The low basket depth suggests transactional purchase behaviour where customers buy specific products rather than multiple items in a single order.

This pattern may indicate:

- limited cross-selling effectiveness  
- absence of strong bundling strategies  
- price-sensitive purchase decisions  

While promotional campaigns successfully increase order participation and revenue contribution, they do not significantly encourage customers to expand their basket size.

Improving bundle offers, recommendation systems, and tiered discount strategies could help increase basket strength and improve overall revenue efficiency.

🟢 Customer Behaviour KPIs

 🎯  KPI = Repeat Purchase Rate

````python
customer_orders = df_net.groupby("Order ID")["Order ID"].count()

repeat_orders = customer_orders[customer_orders > 1].count()

repeat_purchase_rate = ((repeat_orders / total_orders)*100 ).round(3)

print("Repeat Purchase Rate:",repeat_purchase_rate,"%")
````

## 👥 Repeat Purchase Rate Analysis

Customer retention strength was evaluated using the repeat purchase rate metric.

### 📊 Result

- **Repeat Purchase Rate = 5.837%**

This indicates that only a small proportion of customers placed multiple orders during the observed period.

---

### 📌 Key Business Insight

A low repeat purchase rate suggests that the platform is primarily driven by new customer acquisition rather than customer loyalty.

This behaviour may be influenced by:

- heavy promotional dependency  
- category-specific transactional buying  
- limited post-purchase engagement strategies  

Improving customer experience, personalized recommendations, and targeted retention campaigns could help increase repeat purchase behaviour and enhance long-term customer value.
